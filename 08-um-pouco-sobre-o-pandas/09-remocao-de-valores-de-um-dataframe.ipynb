## Remoção de dados de um DataFrame

```python
import pandas as pd 

BASE = "../datasets/Pokemon.csv"

df = pd.read_csv(BASE)
```

```python
df
```

### Apagando coluna(s)

```python
df_without_type2 = df.drop(columns=["type2"])

df_without_type2.head()
```

### Apagando valor(es)

Para apagar valores nós podemos criar um novo DataFrame onde os dados indesejados sejam removidos através de uma condicional.

```python
df_without_invalid_values = df[( df["type1"] != "Blastoise" ) &  ( df["type1"] != "Graass" ) ]
```

Quando executamos a linha acima há um problema os índices do nosso DataFrame ficam com a ordenação comprometida, podemos ver isso em uma nova consulta.

```python
df[ df["type1"] == "Blastoise" ]

# Ao consultar o nosso DataFrame original podemos verificar que um dos valores removidos estava no indice 15
```

```python
df_without_invalid_values[10:16]

# Ao fazer uma consulta no DataFrame que removemos os valores podemos verificar que ele pula do indice 14 para o 16 ignorando o 15 que removemos.
```

Para organizar isso podemos utilizar o `.reset_index()`

```python
df_without_invalid_values = df_without_invalid_values.reset_index()

df_without_invalid_values[10:16]
```
