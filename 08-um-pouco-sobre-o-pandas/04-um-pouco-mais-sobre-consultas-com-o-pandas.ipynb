```python
import pandas as pd

BASE = "../datasets/Pokemon.csv"
```

```python
df = pd.read_csv(BASE)
```

## Acessando multiplas colunas

```python
df[["name", "legendary"]]
```

## Acessando linhas por indices (usando o fatiamento)

```python
df[0:7]
```

## .iloc()

Com o .iloc podemos acessar as linhas do nosso DataFrame por indices

```python
df.iloc[0:5] # Acessando as linhas entre 0 e 4 do nosso DataFrame
```

```python
df.iloc[0:5, 0:3] # Acessando as lihas entre 0 e 4 e pegando as colunas entre cujo os indices estão entre 0 e 2 
```

## .loc()

Com o .loc() conseguimos acessar os registros pelo nome

```python
df.loc[0:5, ["name"]] # Registros cujo o nome das linhas estejam entre as linhas com o nome 0 e a linha com o nome 5
```

## Filtrando registro do nosso DataFrame com critérios lógicos

```python
df[df["sp_attack"] > 70] # Selecionando pokemnos cujo o poder de ataque seja maior do que 70
```

### Com base no retorno acima podemos pegar algumas colunas em especifico

```python
df[df["sp_attack"] > 70][["name", "sp_attack"]]
```

## Filtrando com multiplos critérios

### operador "and" que se torna "&" 

Recomenda-se o uso de parenteses entre cada condição

```python
df[(df["sp_attack"] > 70) & (df["legendary"] == True)]

# Apenas pokemons com sp_atack maior do que 70 e que sejam lendários
```

### operador "or" que se torna "|" 

Recomenda-se o uso de parenteses entre cada condição

```python
df[(df["type1"] == "Ice") | (df["speed"] > 100)]

# Apenas pokemonos do tipo gelo ou cuja a velocidade seja maior do que 100
```

### operador "not" que se torna "~" 

```python
df[~(df["type1"] == "Poison")]

#  Apenas pokemons onde o tipo 1 não seja venenoso
```

## isin()

Para pegarmos registros cujo os valores correspondam aos critérios que passamos como argumento para esta função

```python
df[df["type1"].isin(["Grass", "Ice"])]
```
