```python
import pandas as pd

BASE = "../datasets/Pokemon.csv"
```

```python
df = pd.read_csv(BASE)
```

## Acessando multiplas colunas

```python
df[["name", "legendary"]]
```

## Acessando linhas por indices (usando o fatiamento)

```python
df[0:7]
```

## .iloc()

Com o .iloc podemos acessar as linhas do nosso DataFrame por indices

```python
df.iloc[0:5] # Acessando as linhas entre 0 e 4 do nosso DataFrame
```

```python
df.iloc[0:5, 0:3] # Acessando as lihas entre 0 e 4 e pegando as colunas entre cujo os indices estão entre 0 e 2 
```

## .loc()

Com o .loc() conseguimos acessar os registros pelo nome

```python
df.loc[0:5, ["name"]] # Registros cujo o nome das linhas estejam entre as linhas com o nome 0 e a linha com o nome 5
```
