## O .query() é uma forma de manipular os registros do nosso DataFrame usando uma sintaxe bem similar ao SQL

```python
import pandas as pd

BASE = "../datasets/Pokemon.csv"
```

```python
df = pd.read_csv(BASE)
```

```python
df.query("attack > 100")

# O mesmo que df[df["attack"] > 100]
```

```python
df.query("name == 'Gengar'")

# O mesmo que df[df["name"] == "Gengar"]
```

```python
df.query("type1 == 'Ghost' and attack > 100")

# O mesmo que df[(df["type1"] == "Ghost") & (df["attack"] > 100)]
```

```python
df.query("type1 == 'Ghost' and not attack > 100")

#O mesmo que df[(df["type1"] == "Ghost") & ~(df["attack"] > 100)]
```

```python
df.query("type1 == 'Ghost' or speed > 120")

# O mesmo que df[(df["type1"] == "Ghost")  | (df["speed"] > 120 )]
```

```python
df.query("type1 in ['Ghost', 'Ice']")

# O mesmo que df[df["type1"].isin(["Ghost", "Ice"]) ]
```
