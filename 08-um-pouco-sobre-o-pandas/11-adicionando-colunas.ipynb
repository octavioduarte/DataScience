```python
type_and_addition = {
    "Water":        0.1,
    "Normal":       0.07,
    "Grass":        0.08,
    "Bug":          0.04,
    "Psychic":      0.22,
    "Fire":         0.1,
    "Electric":     0.12,
    "Rock":         0.06,
    "Dark":         0.15,
    "Fighting":     0.07,
    "Ghost":        0.18,
    "Poison":       0.18,
    "Ground":       0.07,
    "Dragon":       0.19,
    "Steel":        0.1,
    "Ice":          0.12,
    "Fairy":        0.12,
    "Flying":       0.11,
    "Legendary":    0.3
}

def add_strategy(row):   
    mean_atk_and_def = (row["attack"] + row["defense"]) / 2
    legendary_weight = type_and_addition["Legendary"] if row["legendary"] else 0
    mean_atk_and_def +=  mean_atk_and_def * type_and_addition[row["type1"]]
    mean_atk_and_def =  mean_atk_and_def + (mean_atk_and_def * legendary_weight)
    return mean_atk_and_def
```

```python
df["strategy"] =  df.apply(add_strategy, axis=1)
```
