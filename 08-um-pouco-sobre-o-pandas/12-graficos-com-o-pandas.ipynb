```python
import pandas as pd 

BASE = "../datasets/Pokemon.csv"

df = pd.read_csv(BASE)
```

O Pandas utiliza por "baixo" dos panos o matplotlib para renderizar gráficos, uma forma bem
simples de exibir gráficos é usando a função "plot".

```python
df["type1"].value_counts().plot()

# Por default a função plot exibe um gráfico de linhas, mas podemos ter diferentes tipos através
#do argumento "kind". 

# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html


#(Extraído da documentação)
#The kind of plot to produce:
#‘line’ : line plot (default)
#‘bar’ : vertical bar plot
#‘barh’ : horizontal bar plot
#‘hist’ : histogram
#‘box’ : boxplot
#‘kde’ : Kernel Density Estimation plot
#‘density’ : same as ‘kde’
#‘area’ : area plot
#‘pie’ : pie plot
#‘scatter’ : scatter plot (DataFrame only)
#‘hexbin’ : hexbin plot (DataFrame only)
```

### Gráfico de barras

```python
df["type1"].value_counts().plot(kind="bar")
```

### Gráfico de pizza

```python
df["type1"].value_counts().plot(kind="pie")
```

### Gráfico de linhas na horizontal

```python
df["type1"].value_counts().plot(kind="barh")
```

### Podemos customizar nosso gráficos com argumentos para a função plot

```python
df["type1"].value_counts().plot(
    kind="bar",
    title="Total de pokemons por tipo", # Título do nosso gráfico
    xlabel="Tipo de pokemon", # Texto do eixo x (eixo horizontal)
    ylabel="Quantidade" # Texto do eixo y (eixo vertical)
)
```
