```python
import pandas as pd 

BASE = "../datasets/HR-Employee-Attrition.csv"

df = pd.read_csv(BASE)
```

#### Criando um contexto para otimizar a visualização da nossa base:

```python
with pd.option_context("float_format", "{:.2f}".format, "display.max_columns", 35):
    display(df.describe())
```

Um ponto interessante é observarmos os dados correspondentes a linha "std" que é o desvio padrão. Podemos ver qual algumas colunas tem o desvio padrão como 0, ou seja, a dispersão dos dados não varia (colunas "EmployeeCount" e "StandardHours"), o que em outras palavras significa que temos o mesmo valor para todos os registros daquela coluna. Este tipo de dado na grande maioria do caso (para não dizer todas) não costuma agregar em nada para nossa análise. São colunas que podemos remover do nosso DataFrame.
