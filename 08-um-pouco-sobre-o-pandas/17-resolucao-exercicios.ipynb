```python
import pandas as pd

BASE = '../datasets/HR-Employee-Attrition.csv'

df = pd.read_csv(BASE)
```

```python
with pd.option_context("float_format", "{:.2f}".format, "display.max_columns", 35):
    display(df.describe())
```

### Remover colunas com valores únicos

```python
for c in df.columns:
    column = df[c]
    if (column.nunique() <= 1):
        print("Removendo coluna: ", c)
        df = df.drop(c, axis=1)
        
```
