```python
import pandas as pd

BASE = '../datasets/HR-Employee-Attrition.csv'

df = pd.read_csv(BASE)
```

### Encontrar registros dos 10 funcionários mais velhos

```python
df.nlargest(10, "Age")
```

### Encontrar registros dos 10 funcionários mais novos¶

```python
df.nsmallest(10, "Age")
```

### Encontrar registros dos 10 funcionários com mais tempo no mesmo cargo¶

```python
df.nlargest(10, "YearsInCurrentRole")
```

### Encontrar registros dos 10 funcionários com menos tempo no mesmo cargo

```python
df.nsmallest(10, "YearsInCurrentRole")
```
